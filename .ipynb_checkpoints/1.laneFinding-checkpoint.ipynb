```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
# read the image & printing stats
image=mpimg.imread("datasets/test.jpg")
print('The type of the image:\n',type(image),'\nThe dim of the image:',image.shape)
plt.imshow(image)
plt.colorbar()
plt.show()
```

```python
# hist
plt.hist(image.ravel(), bins=256, range=(0.0, 1.0), fc='k', ec='k')
plt.show()
```

```python
#take a deep copy
y_size =image.shape[0]
x_size =image.shape[1] 
print('The y & x sizes',y_size, x_size)
color_select=np.copy(image)
```

```python
# defining the color selection critereon:
red_threshold = 200
green_threshold =200
blue_threshold= 200
rgb_threshold = [red_threshold,green_threshold,blue_threshold]

# collect all true/false values depending upon condition
thresholds=   (image[:,:,0] < rgb_threshold[0]) \
            | (image[:,:,1] < rgb_threshold[1]) \
            | (image[:,:,2] < rgb_threshold[2])
print('thresholds shape',thresholds.shape)
color_select[thresholds]=[0,0,0]
print('color_select shape',color_select.shape)
plt.imshow(color_select)
plt.show()
import cv2
cv2.imwrite("datasets/test-after.jpg", color_select)
```
