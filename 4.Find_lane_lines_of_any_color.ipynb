```python
# convert to a gray scale
import cv2  #bringing in OpenCV libraries
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY) #grayscale conversion
plt.imshow(gray, cmap='gray')
plt.show()
print('Shape of image: ',gray.dtype)
```

### Apply [Gauassian smoothing](https://docs.opencv.org/2.4/modules/imgproc/doc/filtering.html?highlight=gaussianblur#gaussianblur)
> 
* a way of suppressing noise and spurious gradients by averaging
* always select a odd number size.. **why?**
<div class="alert alert-block alert-info"><span style='color:red'> &#10003;cv2.Canny() actually applies Gaussian smoothing internally, but we include it here because you can get a different result by applying further smoothing (and it's not a changeable parameter within cv2.Canny()!.</span></div>

```python
# defining a Kernel size for Gaussian smoothing / blurring ?

kernel_size = 5
blur_gray = cv2.GaussianBlur(gray,(kernel_size, kernel_size), 0)
cv2.imwrite("datasets/exit-ramp_gray_GaussianBlur.jpg",blur_gray)
plt.imshow(blur_gray)
plt.show()
print('Shape of image: ',blur_gray.dtype)
```

```python
# Define parameters for Canny and run it
low_threshold = 70
high_threshold = 140
edges = cv2.Canny(blur_gray, low_threshold, high_threshold)
print(edges.shape)
# cv2.imwrite("datasets/exit-ramp_canny_edges.jpg",blur_gray) #will not work as cv2 does not have a color map
mpimg.imsave("datasets/exit-ramp_canny_edges.jpg", arr=edges, cmap='Greys_r')
# Display the image
plt.imshow(edges, cmap='Greys_r')
plt.show()
print('Shape of edges image: ',edges.dtype)
```
